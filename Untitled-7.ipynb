```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, average_precision_score, precision_recall_curve, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('test_hsbc_df.csv')  
```

```python
cols_to_drop = ['zipcodeOri', 'zipMerchant']
existing = [c for c in cols_to_drop if c in df.columns]
if existing:
    df = df.drop(columns=existing)
```

```python
df['amount'] = df['amount'].astype('float32')

categorical_columns = df.select_dtypes(include=['object', 'category']).columns
```

```python

X = df.drop(columns=['fraud']) 
y = df['fraud']
```

```python
# Phase 5: Optional time-based split using `step` if present
# Set this flag to True to use chronological split; otherwise uses stratified random split
USE_TIME_SPLIT = True
TEST_SIZE = 0.30

if USE_TIME_SPLIT and ('step' in X.columns):
    # Sort by step to preserve time order
    df_sorted = df.sort_values('step').reset_index(drop=True)
    X_sorted = df_sorted.drop(columns=['fraud'])
    y_sorted = df_sorted['fraud']
    split_idx = int(len(df_sorted) * (1 - TEST_SIZE))
    X_train, X_test = X_sorted.iloc[:split_idx], X_sorted.iloc[split_idx:]
    y_train, y_test = y_sorted.iloc[:split_idx], y_sorted.iloc[split_idx:]
else:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=TEST_SIZE, random_state=42, stratify=y
    )
```

```python
# Phase 4: Feature grouping and updated Pipeline with OneHot for low-cardinality categoricals
# Determine low-cardinality categorical-like features on the training split
nunique = X_train.nunique()
low_cardinality_cats = [c for c in X_train.columns if (X_train[c].dtype in ['int64', 'int32', 'object', 'category']) and (nunique[c] > 1) and (nunique[c] <= 50)]

# Continuous/numeric features to scale (keep 'amount' and other continuous ints like 'step', 'age' if present)
continuous_candidates = ['amount', 'step', 'age']
numeric_features = [c for c in continuous_candidates if c in X_train.columns]

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numeric_features),
    ('cat_low', OneHotEncoder(handle_unknown='ignore', sparse=False), low_cardinality_cats)
], remainder='passthrough')

pipeline = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('model', LogisticRegression(class_weight='balanced', max_iter=2000, random_state=42))
])

pipeline.fit(X_train, y_train)

print('Low-cardinality categorical features:', low_cardinality_cats)
print('Numeric features (scaled):', numeric_features)
```

```python
# Evaluate pipeline: ROC-AUC, PR-AUC, and threshold tuning via F1
y_proba = pipeline.predict_proba(X_test)[:, 1]

roc = roc_auc_score(y_test, y_proba)
ap = average_precision_score(y_test, y_proba)

prec, rec, thr = precision_recall_curve(y_test, y_proba)
f1_scores = (2 * prec * rec) / (prec + rec + 1e-12)
# precision_recall_curve returns one more precision/recall than thresholds
# Align by ignoring the last point for threshold selection
best_idx = f1_scores[:-1].argmax()
best_threshold = thr[best_idx]

y_pred = (y_proba >= best_threshold).astype(int)
cm = confusion_matrix(y_test, y_pred)
TP = cm[1,1]; FP = cm[0,1]; TN = cm[0,0]; FN = cm[1,0]
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"ROC-AUC: {roc:.4f}")
print(f"PR-AUC (Average Precision): {ap:.4f}")
print(f"Best threshold by F1: {best_threshold:.4f}")
print(f"TP={TP}, FP={FP}, TN={TN}, FN={FN}")
print(f"Precision={precision*100:.2f}%, Recall={recall*100:.2f}%, F1={f1*100:.2f}%")

# Plot ROC and PR curves
from sklearn.metrics import roc_curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(fpr, tpr, label=f'ROC-AUC={roc:.3f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('FPR'); plt.ylabel('TPR'); plt.title('ROC Curve'); plt.legend()

plt.subplot(1,2,2)
plt.plot(rec, prec, label=f'PR-AUC={ap:.3f}')
plt.xlabel('Recall'); plt.ylabel('Precision'); plt.title('Precision-Recall Curve'); plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Phase 4: Business-driven threshold selection (e.g., target precision >= 90%)
# Sweep thresholds to find the highest recall with precision >= target
import numpy as np

target_precision = 0.90
prec, rec, thr = precision_recall_curve(y_test, y_proba)
# Align sizes: prec/rec len = len(thr)+1; use thresholds' indices
valid = []
for i, t in enumerate(thr):
    p = prec[i]
    r = rec[i]
    if p >= target_precision:
        valid.append((t, p, r))

if valid:
    # choose threshold that maximizes recall under precision constraint
    best_t, best_p, best_r = max(valid, key=lambda x: x[2])
else:
    # fallback to best F1 threshold if constraint not achievable
    best_t = best_threshold
    # recompute p/r at fallback
    y_tmp = (y_proba >= best_t).astype(int)
    best_p = precision_score(y_test, y_tmp)
    best_r = recall_score(y_test, y_tmp)

# Evaluate at chosen threshold
y_pred_business = (y_proba >= best_t).astype(int)
cm_biz = confusion_matrix(y_test, y_pred_business)
TPb = cm_biz[1,1]; FPb = cm_biz[0,1]; TNb = cm_biz[0,0]; FNb = cm_biz[1,0]
fbiz = f1_score(y_test, y_pred_business)

print(f"Target precision: {target_precision*100:.0f}%")
print(f"Chosen threshold: {best_t:.4f}")
print(f"Precision={best_p*100:.2f}%, Recall={best_r*100:.2f}%, F1={fbiz*100:.2f}%")
print(f"TP={TPb}, FP={FPb}, TN={TNb}, FN={FNb}")
```

```python
# Phase 4: Stratified 5-fold CV with PR-AUC (Average Precision)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(pipeline, X, y, cv=skf, scoring='average_precision', n_jobs=None)
print('PR-AUC (AP) CV scores:', [f"{s:.4f}" for s in cv_scores])
print(f"PR-AUC (AP) CV mean: {cv_scores.mean():.4f} Â± {cv_scores.std():.4f}")
```

```python
# Phase 6: Hyperparameter tuning (GridSearchCV) for LogisticRegression within the Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold

param_grid = {
    'model__C': [0.01, 0.1, 1.0, 10.0],
    'model__solver': ['lbfgs', 'saga']
}
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    scoring='average_precision',
    cv=cv,
    n_jobs=None,
    refit=True,
    verbose=0
)

grid.fit(X_train, y_train)

pipeline = grid.best_estimator_
print('Best params:', grid.best_params_)
print(f'Best CV AP: {grid.best_score_:.4f}')

# Quick eval on test with tuned pipeline
y_proba = pipeline.predict_proba(X_test)[:, 1]
roc = roc_auc_score(y_test, y_proba)
ap = average_precision_score(y_test, y_proba)
print(f"[Tuned] ROC-AUC: {roc:.4f} | PR-AUC (AP): {ap:.4f}")
```
